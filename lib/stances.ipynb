```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

def stance_distance_analysis(description, support, opposition, model_name = "all-MiniLM-L6-v2"):
    model = SentenceTransformer(model_name)
    embeddings = model.encode([description, support, opposition])

    # Cosine similarities
    sim_pro = float(cosine_similarity([embeddings[0]], [embeddings[1]])[0][0])
    sim_opp = float(cosine_similarity([embeddings[0]], [embeddings[2]])[0][0])
    sim_opp_pro = float(cosine_similarity([embeddings[1]], [embeddings[2]])[0][0])

    # Stance vector norm: |d_pro - d_opp|
    stance_norm = abs((-sim_pro) - (-sim_opp))


    return {
        "similarity_pro": sim_pro,
        "similarity_opp": sim_opp,
        "similarity_opp_pro": sim_opp_pro,
        "stance_norm": stance_norm,
    }
    
description = "Conversion practices ban (sexual orientation)"
#support= "Law should allow conversion or reparative practices aimed at changing a minors sexual orientation."
support= "I love my cat"
opposition= "Law should prohibit conversion or reparative practices aimed at changing a minors sexual orientation."

result = stance_distance_analysis(description, support, opposition)
print(result)
```
